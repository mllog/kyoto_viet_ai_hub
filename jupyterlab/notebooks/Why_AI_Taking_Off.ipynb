# Q. Tại sao gần đây AI/Deep Learning lại mang lại kết quả rất tôt?

<img src="figs/ai_data_performance.png" />

## Three reasons:

- Data: More training data (Big data)

- Computation (e.g More powerfull computation: GPU)

- Algorithms (e.g. More effective activations: ReLu)

## Chu trình thông thương của Deep Learning Projects

<img src="figs/ai_project_cycle.png" />

- Việc quay vòng qui trình từ idea, implement, thử nghiêm nhanh chóng sẽ giúp chúng ta có được kết qủa tốt trong thời gian ngăn hơn.

## GPU (Graphic Processor Unit) giúp tăng cường khả năng tính toán, đăc biệt cho những phép tính dùng trong Deep Learning. 

Nếu chúng ta đang làm những dự án liên quan đến Deep Learning. Việc sử dụng GPU là rất quan trọng để đẩy nhanh quá trình traning models.
